```python
import json
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

dataset_path = '../ChartQA-Dataset/train'
human_file = os.path.join(dataset_path, 'train_human.json')
augumented = os.path.join(dataset_path, 'train_augumented.json')

images_dir = os.path.join(dataset_path, 'png')


with open(human_file, 'r') as f:
    human = json.load(f)

with open(augumented, 'r') as f:
    augument = json.load(f)

print(human)
```

```python
def get_qa_for_image(data, target_image_name):
    """
    Retrieves all questions and answers associated with a specific image name.

    Args:
    data (list): The list of dictionaries containing the dataset.
    target_image_name (str): The filename of the image to search for (e.g., '10095.png').

    Returns:
    str: A formatted string containing the questions and answers.
    """
    # Filter the list for the specific image
    relevant_entries = [entry for entry in data if entry['imgname'] == target_image_name]

    if not relevant_entries:
        return f"No entries found for image: {target_image_name}"

    # Build the readable output
    output = [f"Questions & Answers for {target_image_name}:\n" + "-"*40]
    
    for i, entry in enumerate(relevant_entries, 1):
        question = entry.get('query', 'N/A')
        answer = entry.get('label', 'N/A')
        output.append(f"Q{i}: {question}")
        output.append(f"A{i}: {answer}\n")

    return "\n".join(output)

image_name = '10149.png'
readable_qa = get_qa_for_image(human, image_name)
print(readable_qa)


readable_qa = get_qa_for_image(augument, image_name)
print(readable_qa)
```
